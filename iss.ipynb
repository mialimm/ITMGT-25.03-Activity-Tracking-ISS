```python
import requests
import matplotlib.pyplot as plt
import numpy as np
import time

def iss_loc():
    iss=requests.get("http://api.open-notify.org/iss-now.json")
    loc=iss.json()
    latitude=float(loc["iss_position"]["latitude"])
    longitude=float(loc["iss_position"]["longitude"])
    return longitude, latitude

long=[]
lat=[]

for i in range(100):
    longitude, latitude = iss_loc()
    long.append(longitude)
    lat.append(latitude)
    time.sleep(10.0)
    
x=np.array(long)
y=np.array(lat)

plt.scatter(x, y)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("International Space Station Location")
plt.show()
```
